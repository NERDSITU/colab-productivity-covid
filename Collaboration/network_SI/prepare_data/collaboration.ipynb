```python
import pandas as pd 
from tqdm import tqdm
import networkx as nx
from functools import reduce
```

```python
# now process further to create collaboration
def get_collaboration(df, first, second, outpath): 
  df_pre = df[(df["Date"] >= first) & (df["Date"] < second)] 
  df_pre = df_pre[["PaperId", "AuthorId"]].drop_duplicates() # to be sure
  collaboration_pre = df_pre.merge(df_pre, how = "inner", on = "PaperId")
  collaboration_clean = collaboration_pre[collaboration_pre["AuthorId_x"] < collaboration_pre["AuthorId_y"]]
  collaboration_clean.to_csv(f"{outpath}preprints_{first}_{second}.csv", index = False)
```

```python
# data (from preprocessing)
df_preprints = pd.read_csv(f"{path}/DATA/collaboration/network_SI/preprocessing/paaaa_collaboration.csv")
```

```python
# set outpath
outpath = f"{path}/DATA/collaboration/network_SI/collaboration/"
```

```python
# run collaboration
early = pd.period_range("2010-01-01", "2016-07-01", freq='M')
late = pd.period_range("2015-01-01", "2021-07-01", freq='M')
for early, late in zip(early, late):  
  print('science', str(early), str(late))
  collaboration_df = get_collaboration(df_preprints, str(early), str(late), outpath)
```
