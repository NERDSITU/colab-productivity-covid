```python
# function to create subset dataframe # 
def get_subset(path, filename_in, filename_out): 

  # load file 
  df_main = pd.read_csv(f"{path}{filename_in}")

  # correct dates 
  df_main['month'] = pd.DatetimeIndex(df_main['start_date']).to_period('M') 
  timedelta = 59
  df_main = df_main.assign(month = lambda x: x.month + timedelta)

  # only 2019 --> 
  df_main = df_main[df_main["month"] >= '2019-01']
  # recode gender 

  df_main['Gender'] = df_main['Gender'].apply(lambda x: 'Male' if x == 0 else('Female' if x == 1 else 'Undefined')) 
  # select (relevant) columns
  df_main = df_main[["AuthorId", 
                     "binned", 
                     "Gender", 
                     "NormalizedName", # only for checking, main analysis is across STEM. 
                     "month", 
                     "eigencentrality_unweighted", # I think this hits the floor, but sanity check. 
                     "eigencentrality_weighted", # just for checking 
                     "eigencentrality_unweighted_scaled", # this is probably the main metric. 
                     "eigencentrality_weighted_scaled", # I think this hits the floor, but sanity check. 
                     "coreness", # probably not used. 
                     "degree_weighted",
                     "degree_unweighted"]].drop_duplicates() # important 
  # write file 
  df_main.to_csv(f"{path}{filename_out}", index = False)
```

```python
# load data # 
path = f"{path}/DATA/collaboration/network/main/"
filename_in = "preprints_author_inclusive_GCC_df.csv"
filename_out = "plot_dataframe.csv"
```

```python
# run it # 
get_subset(path, filename_in, filename_out)
```
